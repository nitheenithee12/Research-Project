```
from network.generators.random_graphs
import erdos_renyi_graph
n = 50    ####set number of nodes
p = 1     # probability of connection always keep as 1
g = erdos_renyi_graph(n, p) #Number of vertices and p- Probability of an edge to be included
print(g.nodes)
print(g.edges)
print(len(g.edges))
```

```
import random
#generate random coordinates for points
#2000,2000 is given beacuse area considered is 2km X 2km
coords = [(random.randint(0,2000), random.randint(0, 2000)) for _ in range(301)]
print(coords)
print(len(coords))
```

```
coords[0]=(1000,1000)
import math
pedge=[]
for i in g.edges:
  p1=i[0]
  p2=i[1]
  d = math.sqrt(((coords[p1][0]-coords[p2][0])**2)+((coords[p1][1]-coords[p2][1])**2))

  if d <520: ##set radio range a

    pedge.append((p1,p2))
print((pedge))
print(len(pedge))
```

```
#pedge is connections possible only with given radio range
print(coords)
print(pedge)
```

```
iimport matplotlib.pyplot as plt
import networkx as nx
import numpy as np
%matplotlib inline
G = nx.Graph()
fig,ax = plt.subplots(figsize=(26,15))
plt.title("WIRELESS SENSOR NETWORK")
###draw graph with pedge and number of nodes
for i in range(0,50):###50 nodes so 0 to 50
  G.add_node(i,pos=coords[i])

edges = pedge
G.add_edges_from(edges)
pos=nx.get_node_attributes(G,'pos')
nx.draw(G, pos, with_labels = True, edge_color = 'b',node_size=500,node_color='r',ax=ax,arrows=True)
nx.draw_networkx_edges(G, pos, edge_color='b', arrows = True)
nx.draw_networkx_labels(G, pos)

ax.tick_params(which="both",left=True, bottom=True, labelleft=True, labelbottom=True)
```
