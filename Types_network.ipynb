```python
#@title Default title text
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

k = nx.Graph()

n = 20 #Number of nodes
k.add_nodes_from(range(1,n+1))

for i in range(1,n):
  for j in range(1,n-1):
    if i == j:
      pass
    else:
      k.add_edge(i,j)



nx.draw_random(k, with_labels = True) #draw_random is used to randamize the network.

nm = list(k.nodes)

print(nm)

edges = list(k.edges)
#edges.append((0,0))
print(edges)

print(k.number_of_edges())

print(k.number_of_nodes())



```

```python
import random
coords = [(random.randint(0,5000), random.randint(0, 5000)) for _ in range(20)] # Making coordinates for the given nodes.
print(coords)
print(len(coords))
```

```python
for i in range(0,n): #defining number of nodes to be linked with the coordinates.
  k.add_node(nm[i],pos=coords[i])
```

```python

fig,ax = plt.subplots(figsize=(26,15))
pos=nx.get_node_attributes(k,'pos')
nx.draw(k, pos, with_labels = True, edge_color = 'b',node_size=500,node_color='violet',ax=ax,arrows=True)
print(k.number_of_edges())
```

```python
coords[0]=(1000,1000)
import math
pedge=[]
for i in edges:
  #print(i)
  p1=i[0]
  p2=i[1]
  d = math.sqrt(((coords[p1][0]-coords[p2][0])**2)+((coords[p1][1]-coords[p2][1])**2))

  if d < 1000: ##set radio range a
    pedge.append((p1,p2))
print((pedge))
print(len(pedge))
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
%matplotlib inline
G = nx.Graph()
fig,ax = plt.subplots(figsize=(26,15))
plt.title("WIRELESS SENSOR NETWORK")
###draw graph with pedge and number of nodes
for i in range(0,20):###50 nodes so 0 to 50
  G.add_node(nm[i],pos=coords[i])

edges = pedge
G.add_edges_from(edges)
pos=nx.get_node_attributes(G,'pos')
nx.draw(G, pos, with_labels = True, edge_color = 'b',node_size=500,node_color='r',ax=ax,arrows=True)
nx.draw_networkx_edges(G, pos, edge_color='b', arrows = True)
nx.draw_networkx_labels(G, pos)

ax.tick_params(which="both",left=True, bottom=True, labelleft=True, labelbottom=True)
```
